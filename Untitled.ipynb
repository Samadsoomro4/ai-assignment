```python
class Person:
    def __init__(self, name, age):
        self.name = name
        self.age = age
    
    def greet(self):
        return "Hello, my name is " + self.name + " and I am " + str(self.age) + " years old."

# Creating an instance of the Person class
person1 = Person("Bob", 30)
print(person1.greet())
```
